```python
def linearSolve(a, b):
    '''
        Resolve um sistema linear utilizando o metodo de
        diagonalizacao da matriz (fatoracao de Gauss com pivoteamento) seguido da
        resolucao do sistema triangular superior.
        
        IMPORTANTE (1): Idealmente, trabalhe com floats (coloca um '.' dps do numero inteiro.)
        IMPORTANTE(2): a e b sao arrays, nao listas.
    '''
    
     # ----------- Diagonalizacao ------------- # 
    for i in range(a.shape[0]-1):
        max = i+ np.argmax(np.absolute(a[i:, i])) # Encontra o valor absoluto maximo da coluna
        
        # Alterna duas linhas da array - fase de pivoteamento
        a[[i,max]] = a[[max,i]]
        b[[i,max]] = b[[max,i]]

        # Zera os valores da coluna nas linhas abaixo do pivo atual
        for j in range(i+1, a.shape[0]): 
            m = a[j,i]/(float)(a[i,i])
            a[j] = a[j] - (m*a[i]) 
            b[j] = b[j] - (m*b[i])
      
    # ----------- Resolucao do sistema triangular inferior ------------- # 
    x = np.empty(b.shape)
    for i in np.arange(a.shape[0]-1, -1, -1): # i percorre n, n-1, n-2 ateh 0.
        # Descobre o valor do x[i]
        x[i] = b[i]
        for j in np.arange(a.shape[0]-1, i, -1):
            x[i] -= (a[i, j] * x[j])
        x[i] /= (float)(a[i,i])
    return x
```

```python
# Teste da funcao linearsolve().
a = np.array([[.3e33,1., .5e4], [1.,2., .5e4],[3.5, 4., .5e-4]])
b = np.array([9. ,8., .1])

x = linearSolve(a, b)
print x
print "testando solucao..."
if np.allclose(np.dot(a, x), b):    # Verifica se a multiplicacao de matriz A*x resultou em b.
    print "Sistema resolvido corretamente."
else:  # Caso haja divergencia...
    print "Resposta esperada:", np.linalg.solve(a, b) # Compara com a funcao de resolucao da biblioteca de Python
```

```python
def gaussSeidel(a, b, k_max=1000000, tolerancia=1e-5):
    '''
        Resolve um sistema linear utilizando o metodo de
        Gauss-Seidel.
        
        IMPORTANTE (1): Idealmente, trabalhe com floats (coloca um '.' dps do numero inteiro.)
        IMPORTANTE (2): a e b sao arrays, nao listas.
        IMPORTANTE (3): A resposta pode divergir. Teste a matriz pelo criterio de Sassefeld para ter certeza.
    '''
    
    x = np.ones(len(b)) #Define chute inicial (tudo 1) 
    k = 0 #Contagem de iteracoes
    erro = tolerancia+1 #Define erro
    
    # Enquanto as condicoes de parada nao forem satisfeitas
    while k < k_max and erro > tolerancia:
        x_0 = np.copy(x) #Copia para verificacao do erro relativo
        
        
        #Atualiza os elementos de x
        for i in range(len(x)):
            x[i] = b[i]
            for j in range(len(x)):
                if i != j:
                    x[i] -= (a[i,j] * x[j])
            x[i] /= a[i,i]
        
        k+=1    #Atualiza o num de iteracoes....
        erro = np.max(np.absolute(np.subtract(x,x_0))) / np.max(np.absolute(x)) # e atualiza o erro.
    
    return x
```

```python
# Teste da funcao gaussSeidel().
a = np.array([[2., 1.],[3., 4.]])
b = np.array([1., -1.])

x = gaussSeidel(a, b)
print x
print "testando solucao..."
if np.allclose(np.dot(a, x), b):    # Verifica se a multiplicacao de matriz A*x resultou em b.
    print "Sistema resolvido corretamente."
else:
    print "Resposta esperada:", np.linalg.solve(a, b)   # Compara com a funcao de resolucao da biblioteca de Python
```

## Aproximações para a a tabela por mínimos quadrados

```python
def residuoMMQ(f, x, y):
    '''
        Calcula o residuo (soma dos quadrados dos erros) da aprox f de y nos pontos x.
    '''
    soma = 0
    for xk, yk in zip(x,y):
        soma += pow(f(xk) - yk, 2)
    return soma
```

### Função linear

Aproximaremos a tabela acima por uma função no modelo
$$f(x) = a \cdot g_1(x) + b \cdot g_2(x),$$
com $g_1(x) = x$ e $g_2(x) = 1$.
Para descobrir os coeficientes, queremos que a função 
$$L(a, b) = \sum\limits_{k=1}^m (f(x_k) - y_k)^2$$
seja mínima. para isso, $\frac{\partial L(a)}{\partial a} = 0$ e $\frac{\partial L(b)}{\partial b} = 0$. Realizada as derivações, chegamos no seguinte sistema de equações a ser resolvido:

$$\sum\limits_{k=1}^m x_k^2 \cdot a + \sum\limits_{k=1}^m x_k \cdot b = \sum\limits_{k=1}^m (x_k \cdot y_k) $$
$$\sum\limits_{k=1}^m x_k \cdot a + \sum\limits_{k=1}^m 1 \cdot b = \sum\limits_{k=1}^m (1 \cdot y_k) $$

```python
def linearMMQ(x, y, verificacao=True):
    '''
        Aproxima pontos dados numa funcao linear pelo metodo dos minimos quadrados.
    '''
    # -------------- Encontra os coeficientes das matrizes do sistema ---------------- #
    somaX = np.sum(x)
    somaQuadradoX = np.sum(np.square(x))
    somaY = np.sum(y)
    somaXY = np.sum(np.multiply(x, y))
    num = len(y)

    a = np.array([[somaQuadradoX, somaX],
                  [somaX, num]])
    b = np.array([somaXY, somaY])
    
    
     # ---------- Resolve o sistema linear ---------- #
    print "Resolvendo o sistema linear A * c = B:"
    print "A =", a
    print "B =", b
    
    c = linearSolve(a, b) # Resolve o sistema linear
    
    if (verificacao):
        print "testando solucao..."
        if np.allclose(np.dot(a, c), b):
            print "Sistema resolvido corretamente."
        else:
            print "Resposta esperada:", np.linalg.solve(a, b)
        return c
```

```python
c1 = linearMMQ(diaArray, niveisArray)
```

```python
aproxLin = lambda x: c1[0]*x + c1[1] #funcao linear encontrada

linearY = np.vectorize(aproxLin)(dia)

plt.plot(dia, niveis, 'b-')
plt.plot(dia, linearY, 'r-')
plt.title("Nivel da Usina Cantareira entre 2003 e 2015")
plt.ylabel('nivel da usina')
plt.xlabel("dia")
plt.show()
print "O residuo encontrado na aproximacao foi:",residuoMMQ(aproxLin, dia, niveis)
```

### Função de 4o grau

Para fazer a aproximação, consideramos o modelo


$$f(x) = a \cdot g_1(x) + b \cdot g_2(x) + c \cdot g_3(x) + d \cdot g_4(x) + e \cdot g_5(x),$$


com $g_1(x) = x^4$, $g_2(x) = x^3$, $g_3(x) = x^2$, $g_4(x) = x$ e $g_5(x) = 1$.

Realizando o mesmo processo descrito anteriormente para a aproximação linear, chegamos no sistema:


$$\sum\limits_{k=1}^m x_k^8 \cdot a + \sum\limits_{k=1}^m x_k^7 \cdot b + \sum\limits_{k=1}^m x_k^6 \cdot c + \sum\limits_{k=1}^m x_k^5 \cdot d + \sum\limits_{k=1}^m x_k^4 \cdot e= \sum\limits_{k=1}^m (x_k^4 \cdot y_k) $$
$$\sum\limits_{k=1}^m x_k^7 \cdot a + \sum\limits_{k=1}^m x_k^6 \cdot b + \sum\limits_{k=1}^m x_k^5 \cdot c + \sum\limits_{k=1}^m x_k^4 \cdot d + \sum\limits_{k=1}^m x_k^3 \cdot e= \sum\limits_{k=1}^m (x_k^3 \cdot y_k) $$
$$\sum\limits_{k=1}^m x_k^6 \cdot a + \sum\limits_{k=1}^m x_k^5 \cdot b + \sum\limits_{k=1}^m x_k^4 \cdot c + \sum\limits_{k=1}^m x_k^3 \cdot d + \sum\limits_{k=1}^m x_k^2 \cdot e= \sum\limits_{k=1}^m (x_k^2 \cdot y_k) $$
$$\sum\limits_{k=1}^m x_k^5 \cdot a + \sum\limits_{k=1}^m x_k^4 \cdot b + \sum\limits_{k=1}^m x_k^3 \cdot c + \sum\limits_{k=1}^m x_k^2 \cdot d + \sum\limits_{k=1}^m x_k \cdot e= \sum\limits_{k=1}^m (x_k \cdot y_k) $$
$$\sum\limits_{k=1}^m x_k^4 \cdot a + \sum\limits_{k=1}^m x_k^3 \cdot b + \sum\limits_{k=1}^m x_k^2 \cdot c + \sum\limits_{k=1}^m x_k \cdot d + \sum\limits_{k=1}^m 1 \cdot e= \sum\limits_{k=1}^m (1 \cdot y_k) $$

```python
def quartoGrauMMQ(x, y, verificacao=True):
    '''
        Aproxima pontos dados numa funcao polinomial de 4o grau pelo metodo dos minimos quadrados.
    '''
    
    # ---------- Encontra os coeficientes das matrizes do sistema ---------- #
    somaPotX = []
    for i in range(9):
        sum = 0 #Pode ser reduzida para somaPotX.append(np.sum(x**i)) se nao houver overflow
        for xk in x:
            sum += pow(xk, i)
        somaPotX.append(sum)
        
    somaPotXY = []
    for i in range(5): 
        soma = 0 #Pode ser reduzida para somaPotXY.append(np.sum(np.multiply(x**i, y))) se nao houver overflow
        for xk, yk in zip(x,y):
            soma += (pow(xk, i)*yk)
        somaPotXY.append(soma)
    
    # Ajustando os coeficientes das matrizes A e B
    a = np.array(
        [[somaPotX[8], somaPotX[7], somaPotX[6], somaPotX[5], somaPotX[4]],
         [somaPotX[7], somaPotX[6], somaPotX[5], somaPotX[4], somaPotX[3]],
         [somaPotX[6], somaPotX[5], somaPotX[4], somaPotX[3], somaPotX[2]],
         [somaPotX[5], somaPotX[4], somaPotX[3], somaPotX[2], somaPotX[1]],
         [somaPotX[4], somaPotX[3], somaPotX[2], somaPotX[1], somaPotX[0]]])
         
    b = np.array(somaPotXY[::-1]) #B eh a array somaPotXY de baixo pra cima
    
    
    print "Resolvendo o sistema linear A * c = B:"
    print "A =", a
    print "B =", b
    
    # ---------- Resolve o sistema linear ---------- #
    c = linearSolve(a, b)
    
    if (verificacao):
        #testando a solucao
        print "testando solucao..."
        if np.allclose(np.dot(a, c), b):   # Verifica se a multiplicacao de matriz A*x resultou em b.
            print "Sistema resolvido corretamente."
        else:
            print "Resposta esperada:", np.linalg.solve(a, b) # Compara com outra funcao de resolucao
        return c
```

```python
c4 = quartoGrauMMQ(diaArray, niveisArray)
print c4
```

```python
aprox4o = lambda x: c4[0]*x**4 + c4[1]*x**3 + c4[2]*x**2 + c4[3]*x**1 + c4[4] # gera a funcao encontrada

quartoGrauY = np.vectorize(aprox4o)(dia)

plt.plot(dia, niveis, 'b-')
plt.plot(dia, quartoGrauY, 'r-')
plt.title("Nivel da Usina Cantareira entre 2003 e 2015")
plt.ylabel('nivel da usina')
plt.xlabel("dia")
plt.show()
print "O residuo encontrado na aproximacao foi:",residuoMMQ(aprox4o, dia, niveis)
```

### Função polinomial de grau arbitrário

Para testar outras aproximações, foi desenvolvida um método que modela pontos a uma função polinomial de grau arbitrário, e verificou-se qual grau que fornece a melhor aproximação pelo método de "força bruta". Note que, em teoria, quanto maior o grau, a aproximação se torna potencialmente melhor, No entanto, isso não é necessariamente verdade na prática, já que pode haver perda de precisão ao tratar-se altas potências como é o caso:

```python
def grauX_MMQ(x, y, grau=4, verborragic=True):
    '''
        Aproxima pontos dados numa funcao polinomial de grau arbitrário pelo metodo dos minimos quadrados.
        A precisão depende da máquina, sendo seu "BottleNeck" a implementacao do NumPy - que por sua vez facilitou
        bastante a implementacao.
    '''
    
    # ---------- Encontra os coeficientes das matrizes do sistema ---------- #
    somaPotX = []
    for i in range(grau*2+1):
        sum = 0 #Pode ser reduzida para somaPotX.append(np.sum(x**i)) se nao houver overflow
        for xk in x:
            sum += pow(xk, i)
        somaPotX.append(sum)
        
    somaPotXY = []
    for i in range(grau+1): 
        soma = 0 #Pode ser reduzida para somaPotXY.append(np.sum(np.multiply(x**i, y))) se nao houver overflow
        for xk, yk in zip(x,y):
            soma += (pow(xk, i)*yk)
        somaPotXY.append(soma)
    
    # Ajustando os coeficientes das matrizes A e B
    a = np.array([[somaPotX[j] for j in range (2*grau - i, grau-i-1, -1)] for i in range (grau+1)])
         
    b = np.array(somaPotXY[::-1]) #B eh a array somaPotXY de baixo pra cima
    
    # ---------- Resolve o sistema linear ---------- #
    if verborragic:
        print "Resolvendo o sistema linear A * c = B:"
        print "A =", a
        print "B =", b
    
    c = linearSolve(a, b)
    
    if verborragic:
        #testando a solucao
        print "testando solucao..."
        if np.allclose(np.dot(a, c), b):  # Verifica se a multiplicacao de matriz A*x resultou em b.
            print "Sistema resolvido corretamente." 
        else:
            print "Resposta esperada:", np.linalg.solve(a, b) # Compara com outra funcao de resolucao
    return c
```

```python
def aproxGrauX(x, y, grau=4):
    '''
        Funcao de conveniencia que ja retorna automaticamente a funcao-aproximacao.
    '''
    c = grauX_MMQ(x, y, grau, verborragic=False)
    return lambda x: np.sum([c[j]*x**(grau-j) for j in range(grau+1)])
```

```python
# ------------ Forca bruta para encontrar melhor polinomio de aproximacao --------- #
menorRes = 100000000000000.
menorInd = -1
for i in range (15, 30):  # Testa entre graus 15 e 29
    aprox = aproxGrauX(diaArray,niveisArray, grau=i)
    residuo = residuoMMQ(aprox, dia, niveis)
    print "indice", i , "Residuo:", residuo
    if residuo < menorRes: # Verifica se encontrou residuo menor do que o ja existente
        menorRes = residuo
        menorInd = i
print "Menor residuo:", menorRes, "Indice:", menorInd
```

```python
#Testando com o menor indice, 19
c19 = grauX_MMQ(diaArray, niveisArray, 19, verborragic=False)
print c19
```

```python
aprox19o = lambda x: np.sum([c19[j]*x**(19-j) for j in range(20)])
grau19Y = np.vectorize(aprox19o)(dia)

plt.plot(dia, niveis, 'b-')
plt.plot(dia, grau19Y , 'r-')
plt.title("Nivel da Usina Cantareira entre 2003 e 2015, e a melhor aproximacao encontrada")
plt.ylabel('nivel da usina')
plt.xlabel("dia")
plt.show()
print "O residuo encontrado na aproximacao foi:",residuoMMQ(aprox19o, dia, niveis)
```

# Parte 2 - Equação de Calor

Nesta etapa, vamos resolver uma equação diferencial parcial através dos Métodos das Diferenças Finitas. Mais especificamente, trabalharemos com uma equação do calor, que modela a temperatura de uma barra afetada por uma fonte de calor, utiliza duas variáveis, tempo e espaço. Assim, pode ser demonstrado que a função u(x, t) deve ser solução da seguinte equação diferencial

$$∂u/∂t (x, t) = γ∂²u/∂x²(x, t) + ψ(x, t)$$
                                    
onde γ é o coeficiente de condução de calor e ψ(x, t) é uma função que representa a fonte de calor. Além disso, devemos impor condições iniciais e de contorno sobre o problema.

$$u(x, 0) = f(x)$$
$$u(0, t) = p(t)$$
$$u(L, t) = q(t).$$

Este tipo de condição inicial dado é chamado de Condição de Dirichlet. Se, por exemplo, um dos lados da barra estiver isolado e não conduzir calor, trocaremos a condição de contorno correspondente por uma condição do tipo ux(x, t) = 0 (ux é a derivada de u(x, t) com relação a x), chamada de Condição de Neumann. Uma maneira simples de se resolver problemas deste tipo é usando um esquema de diferenças finitas, fazendo discretizações no tempo e no espaço [5]. Para isto, usaremos

$$u(xi, tn) ≈ u(n,i).$$
                                                
________________________________________________________________________________________________________________________________

Para aproximar a derivada temporal usando diferenças avançadas, fazemos

$$ut(xi, tj ) ≈ u(j+1 i) − u(j,i)/k, (2)$$

onde k é o passo associado à variável temporal, e j = 0, 1, . . . .
Agora, a derivada espacial será sempre aproximada por

$$uxx(xi, tj ) ≈ u(j, i+1) − 2u (j, i) + u(j, i−1)/h², (3)$$

onde h é tal que xı = ıh, para ı = 0, 1, . . . , m, sendo m um inteiro positivo.
Assim, podemos substituir (3) e (2) em (1) e obter

$$u(j+1, i) − u(j, i)/k = γu(j, i+1) − 2u(j, i) + u(j, i−1)/h² + ψ(j, i)$$

Podemos, então, explicitar
$$u(j+1, i)$$
                                                          
em termo das outras variáveis, encontrando

$$u(j+1, i) = (1 − 2kγ/h²) u(j, i) + γk/h² . (u(j, i+1) + u(j, i−1)) + kψ(j, i), i = 1, . . . , m, j > 0 (4)$$
        
Simplificando em forma matricial, temos 
                                            
$$u(j+1) = Au(j) + kψ(j)$$
                                                      
Em que A é a matriz tridiagonal de diagonais (lambda, 1 - 2 . lambda, lambda), com (lambda = γk/h²).
Essa é a fórmula que usaremos abaixo no método das diferenças avançadas.

________________________________________________________________________________________________________________________________

Para aproximar a derivada temporal usando diferenças atrasadas, fazemos

$$ut(xi, tj ) ≈ u(j, i) − u(j−1, i)/k, (5)$$

Substituindo (3) e (5) em (1), temos

$$u(j, i) − u(j−1, i)/k = γu(j, i+1) − 2u(j, i) + u(j, i−1)/h² + ψ(j, i)$$

Podemos, então, isolar

$$u(j−1, i)$$

e escrevê-la em termo das outras variáveis, encontrando

$$u(j−1, i) + kψ(j, i) = −γk/h²2 . u(j, i−1) + (1 + 2kγ/h²) u(j, i) − γk/h² . u(j, i+1), i = 1, . . . , m, j > 0 (6)$$
        
Simplificando em forma matricial, temos 
                                            
$$Au(j+1) = u(j) + kψ(j)$$
                                                      
Em que A é a matriz tridiagonal de diagonais (lambda, 1 - 2 . lambda, lambda), com (lambda = γk/h²).
Essa é a fórmula que usaremos abaixo no método das diferenças avançadas.

# Método da Diferenças Finitas Avançadas

Este método utiliza uma multiplicação matricial para encontrar o novo veotor u(j+1) a partir de um vetor u(j). Utilizamos a função de multiplicação matricial do NumPy para este processo. A função base foi fixada por conveniência, mas todos os outros parâmetros são recebidos na entrada

Note que quando o lambda (gamma.k/h²) é maior que 1/2, o método retorna resultados de crescente instabilidade. Contudo, ele é sensivelmente mais rápido caso lambda seja menor que 1/2

```python
import numpy as np
import math

def funcBase(x):
    return math.sin(math.pi*x) + (x-x**2)     #funcao base fixada
```

```python
def tridiag(a, b, c, k1=-1, k2=0, k3=1):
    '''
    Esta função cria uma matriz tridiagonal a partir de três vetores, usando a função np.diag da biblioteca NumPy
    '''
    return np.diag(a, k1) + np.diag(b, k2) + np.diag(c, k3)
```

```python
def diferencasAvancadas2(gamma = 1, L = 1, K = 0.005, H = 0.01, numIter = 5):
    
    '''
    Esta funcao aproxima a solução de uma equação diferencial de segundo grau, uma equação do calor,
    atravéz do método das diferenças finitas avançadas, utilizando a multiplicação de matriz da biblioteca NumPy.
    Este método tende a ser instável, especialmente quando lambda = gamma*k/h^2 >= 1/2 .
    '''

    size = (L/(H))-1                               # inicializa o tamanho da matriz e do vetor de soluções

    uJ = []                                        # inicializa o vetor de solucoes
    saida = []                                     # inicializa o vetor de vetores da saida
    
    for cont in range(1, int(size)+1):             # este 'for' roda size vezes, exceto em zero para inicializar
        uJ.append(funcBase(cont*H))                # 'u(0)', um vetor com "size" elementos, no vetor de solucoes
    saida.append(uJ)                               # insere esta solucao na posicao '0' do vetor de saida
    
    lmbda = gamma*K/(H**2)                         # inicializa o lambda
    
    matrizA = []                                   #
    menosL = []                                    # inicializa os vetores auxiliares para fazer a tridiagonal
    umMaisL = []                                   #
    
    for cont in range(0, int(size)-1):             #
        menosL.append(-1*lmbda)                    # preenche os vetores auxiliares com
    for cont in range(0, int(size)):               # os elementos usados na tridiagonal
        umMaisL.append(1+(2*lmbda))                #
    
    matrizA = tridiag(menosL, umMaisL, menosL)     # faz a matriz tridiagonal usando a funcao tridiag
    
    for contIter in range(0, numIter):            # este 'for' roda para cada iteracao requerida
        uPlus = []                                 # cria/anula o vetor que contem a nova solucao
        
        for cont in range(0, int(size)):           # preenche o vetor da nova solucao com
            uPlus.append(uJ[cont]+2*K)             # valores dados pela formula do metodo
            
        uPlus = np.dot(matrizA, uPlus)             # faz a multiplicacao da matriz com o vetor

        uJ = uPlus                                 # substitui o vetor U(j) pelo novo vetor
        
        saida.append(uJ)                           # insere o vetor de solucao atual no vetor de saida
        
    saidaArray = np.array(saida)
    
    plt.plot( uJ, 'b-')                            #
    plt.title("Solucoes em tempo t = %f"% (K))     # imprime um grafico para
    plt.ylabel('Solucao')                          # o vetor de solucao 'uJ'
    plt.show()
    
    print(saidaArray[numIter-1])                   # imprime o vetor de saida
```

```python
diferencasAvancadas2(1, 1, 0.000001, 0.01, 5)  #lambda = 1/100 < 1/2
```

```python
diferencasAvancadas2(1, 1, 0.00001, 0.01, 5) #lambda = 1/10 < 1/2
```

```python
diferencasAvancadas2(1, 1, 0.00005, 0.01, 5)  #lambda = 1/2 = 1/2
```

```python
diferencasAvancadas2(1, 1, 0.001, 0.01, 5)   #lambda = 10 > 1/2
```

```python
diferencasAvancadas2(1, 1, 0.01, 0.01, 5)   #lambda = 100 > 1/2
```
